# Encoding the categoricals variables

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
data['ShelveLoc'] = LE.fit_transform(data['ShelveLoc'])
data['Urban'] = LE.fit_transform(data['Urban'])
data['US'] = LE.fit_transform(data['US'])
```

```python
data.head()
```

```python
data.Sales.mean(),data.Sales.max()
```

```python
sns.histplot(data.Sales)
```

# Converting Sales column into the categories for the better prediction

```python
bins = [0,5.39,9.32,17]
labels = ['Low',"Mid",'High']
sales_ctg = pd.cut(data['Sales'],bins=bins,labels=labels)
sales_ctg = pd.DataFrame(sales_ctg)
data1 = pd.DataFrame(data.drop(columns='Sales'))
data1['Sales_Cat'] = sales_ctg
data1
```

```python
data1[data1['Sales_Cat'].isnull()]
```

```python
data.corr(numeric_only=True)
```

```python
sns.heatmap(data1.corr(numeric_only=True))
```

# Encoding the Sales column

```python
data2 = pd.DataFrame(data1)
data2['Sales_enc'] = LE.fit_transform(data1.Sales_Cat)
data2.drop('Sales_Cat',axis=1,inplace=True)
data2
```

```python
X = data1.iloc[:,:-2]
```

```python
X
```

```python
Y = data['Sales']
```

# Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y1,test_size=0.2,random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
from sklearn import ensemble
model = ensemble.RandomForestClassifier(random_state=42)
```

# Hyper parameters tuning

```python
parameters = [{'n_estimators':[20,40,60,80,100],'max_depth':[5,10,15,20]},{'n_estimators':[20,30,40,60,100], 'max_depth': 
             [5,10,15],'max_features':[2,5,8,10]}]
```

# Grid Search CV

```python
grid_search = GridSearchCV(estimator=model,param_grid=parameters,n_jobs=-1,cv=10)
```

```python
grid = grid_search.fit(X_train,y_train)
```

```python
grid.best_score_
```

```python
grid.best_params_
# We got best score at max_depth': 15, 'max_features': 10, 'n_estimators': 30
```
