# Create frequent datasets

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

# 1. Association rules with 10% Support and 70% confidence

```python
# With 10% support
freq_books = apriori(book,min_support=0.1,use_colnames=True)
```

```python
freq_books
```

# From frequent itemsets extract the association rules

```python
ass_rules = association_rules(freq_books,metric='lift',min_threshold=0.7)
```

```python
ass_rules.sort_values('lift',ascending=False)
```

```python
## A leverage value of 0 indicates independence. Range will be [-1 1]
## A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]
```

```python
ass_rules.sort_values('lift',ascending=False)[0:20]
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
ass_rules[ass_rules.lift>1]
```

```python
ass_rules[['support','confidence']].hist()
```

```python
ass_rules[['support','confidence','lift']].hist()
```

```python
# visualization of obtained rule
import matplotlib.pyplot as plt

plt.scatter(ass_rules['support'], ass_rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```

```python
import seaborn as sns
sns.scatterplot(x = 'support', y = 'confidence', data=ass_rules, hue='antecedents')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
```
