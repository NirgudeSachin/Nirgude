# Scaling the Data


```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar = MinMaxScaler()
```

```python
scaled_df=scalar.fit_transform(df)
```

```python
scaled_df
#scaled crimedata
```

# Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters=3,linkage='average')
```

```python
hc_pred = hc.fit_predict(scaled_df)
```

```python
hc_pred
```

```python
hc_pred = pd.DataFrame(hc_pred,columns=['Cluster'])
```

```python
hc_pred
```

```python
Hierarchical_clustering = pd.concat([crime_data,hc_pred.Cluster],axis=1)
```

```python
Hierarchical_clustering.head()
```

```python
agglo_grp = Hierarchical_clustering.groupby('Cluster')
```

```python
agglo_grp.agg('mean')
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(scaled_df,hc_pred)
```

```python
import scipy.cluster.hierarchy as sch 
```

```python
sch.linkage(scaled_df,method = 'average')
```

```python
plt.figure(figsize=(12,6))
sch.dendrogram(sch.linkage(scaled_df,optimal_ordering= False,method = 'average'))
plt.show()
```

Hirearchichal -It has been divided into 3 clusters as 0,1, and 2 .

# KMeans Clustering 

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3,random_state=42)
```

```python
km_pred = km.fit_predict(scaled_df)
```

```python
km_pred
```

```python
silhouette_score(scaled_df,km_pred)
```

```python
wcss = []
for i in range(1,15):
    km = KMeans(i,random_state=42)
    km_pred = km.fit_predict(scaled_df)
    wcss.append(km.inertia_)
```

```python
wcss
```

KMeans- it is clear that it has k clusters=3 .... and based on that it has been divide into 3 clusters.

# Elbow Plot

```python
plt.plot(wcss)
```

# DB Scan

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.3,min_samples=5)

```

```python
db_pred = db.fit(scaled_df)
```

```python
a= db_pred.labels_
```

```python
a = pd.DataFrame(a) 
```

```python
a.nunique().sum()
```

```python
a
```

```python
dbscan_df = pd.DataFrame(df)
dbscan_df['Dbscan_df'] = a
```

```python
dbscan_df
```

DBSCAN- Based on the data , its eps and min_samples , model has predicted points on basis of -1(noise points),0(core points),1(border points)
