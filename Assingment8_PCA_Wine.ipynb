# Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
sc_wine = sc.fit_transform(wine)
sc_wine
```

```python
sc_wine = pd.DataFrame(sc_wine,columns=wine.columns)
sc_wine
```

# PCA

```python
from sklearn.decomposition import PCA
pca = PCA()
pc =pca.fit_transform(sc_wine)
pc
```

```python
pc = pca.fit_transform(sc_wine)
pc = pd.DataFrame(pc)
pc
```

```python
pca = PCA(n_components=3)
pca_3 = pca.fit_transform(sc_wine)
pca_3
```

# Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
agc = AgglomerativeClustering(n_clusters=3,metric='euclidean')
```

```python
clus_ag = agc.fit_predict(pca_3)
clus_ag
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(pca_3,clus_ag)
```

```python
hierar_cluster = pd.DataFrame(sc_wine)
hierar_cluster['Heirarchichal_cluster']= clus_ag
hierar_cluster
```

# Dendogram

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram

z = linkage(pca_3,method='ward',metric='euclidean')
```

```python
sch = dendrogram(z)
```

# Kmeans Clustering 

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3)
```

```python
km_cluster = km.fit_predict(pca_3)
km_cluster
```

```python
silhouette_score(pca_3,km_cluster)
```

# The optimum number of clusters are 3

```python
wcss = []
for i in range(1,15):
    kc = KMeans(n_clusters=i,random_state=42)
    kc.fit_predict(pca_3)
    wcss.append(kc.inertia_)
```

```python
wcss
```

# Elbow curve

```python
plt.plot(range(1,15),wcss)
# The optimum number of clusters are 3
```

# Dataframe of Clusters

```python
kmean_cluster = pd.DataFrame(sc_wine)
kmean_cluster['KMeans_Cluster']= km_cluster
kmean_cluster
```
